### Web Scrapping "HoopsHype" for 2019 NBA Free Agents

```python
from requests import get
```

```python
url = 'https://hoopshype.com/2019/02/08/nba-free-agency-2019-the-top-players/'
response = get(url)
```

```python
from bs4 import BeautifulSoup
```

```python
html_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#Getting container for each player with their information.  
player_containers = html_soup.find_all('div', attrs={'style':'border-bottom:1px solid #ea362e;margin-bottom:10px;padding-bottom:10px;'})

#Total of 70 Free Agents in 2019
len(player_containers)
```

```python
#Find all returns values in a list.
first_player = player_containers[0]
```

```python
first_name = first_player.a.text
first_name
```

```python
first_option = first_player.b.text
first_option
```

```python
first_team = first_player.find_all('b')[5].text[6:]
first_team
```

```python
first_agent = first_player.find('div', attrs={'style':'display:inline-block;width:32%;'}).text[7:]
first_agent
```

#### Do the same thing to scrap the values for all the other players with a loop

```python
#Lists to store the scraped values
names = []
options = []
team2018_2019 = []
agent = []
```

```python
#Loop to grab all the free agents
for container in player_containers:
    
    #Get the name for the player and put it in the list created above
    name = container.a.text
    names.append(name)

    #Get the option for the player and put it in the list created above
    option = container.b.text
    options.append(option)

    #Get the team for the player and put it in the list created above
    team = container.find_all('b')[5].text[6:]
    team2018_2019.append(team)

    #Get the agent name for the player and put it in the list created above
    agents = container.find('div', attrs={'style':'display:inline-block;width:32%;'}).text[7:]
    agent.append(agents)
```

```python
#Put it in a dataframe with pandas
import pandas as pd

df = pd.DataFrame({'Player': names,
                  'Option': options,
                  'Team2018_2019': team2018_2019,
                  'Agent': agent})
```

```python
#This is how the final results look
df.head(50)
```

```python
#Export the data as .csv file to be used for a project, "2010-2017 NBA Stats Analyzing to find best basketball shooters for Los Angeles Lakers"
df.to_csv('NBA2019FreeAgents.csv')
```
